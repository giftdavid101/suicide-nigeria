### SCALAR VALUES

```python
#Average suicide rate for both genders from 2000-2019
suicide_rate_for_total_genders = Nigerian_suicide_cases[(Nigerian_suicide_cases["Gender"] == "TOTAL") & (Nigerian_suicide_cases["DIM_AGE"] == "TOTAL")]

#Loop and print suicide value for total gender for each year
for year in sorted(suicide_rate_for_total_genders["Suicide_Year"].unique()):
    suicide_rate = suicide_rate_for_total_genders[suicide_rate_for_total_genders["Suicide_Year"] == year]["Suicide_Rate_N"].values[0]
    print(f"Year: {year}, Suicide Rate: {suicide_rate}")

# Calculate average value from 2000-2019
Average_suicide_value = suicide_rate_for_total_genders["Suicide_Rate_N"].mean()
print(f"\nAverage Suicide Rate from (2000–2019): {Average_suicide_value:.2f}")
```

```python
# Get National average suicide rate 2019
overall_rate = Nigerian_suicide_cases[(Nigerian_suicide_cases["DIM_AGE"] == "TOTAL") & (Nigerian_suicide_cases["Gender"] == "TOTAL")]
latest_year = overall_rate["Suicide_Year"].max()
latest_rate = overall_rate[overall_rate["Suicide_Year"] == latest_year]["Suicide_Rate_N"].values[0]
print(f"The National average suicide rate for the year {latest_year} is: {latest_rate}")
```

```python
# Age group and gender with the most alarming (highest) suicide rate
row = Nigerian_suicide_cases.loc[Nigerian_suicide_cases["Suicide_Rate_N"].idxmax(), ["DIM_AGE", "Gender", "Suicide_Rate_N"]]
print(f"Age group: {row['DIM_AGE']}, Gender: {row['Gender']}, Rate: {row['Suicide_Rate_N']} for the year 2019")
```

```python
# Age group and Female gender suicide rate
female_row = year_2019[year_2019["Gender"] == "FEMALE"].loc[year_2019[year_2019["Gender"] == "FEMALE"]["Suicide_Rate_N"].idxmax(), ["DIM_AGE", "Suicide_Rate_N"]]
print(f"Female — Age group: {female_row['DIM_AGE']}, Rate: {female_row['Suicide_Rate_N']} for the year 2019")
```

# Visualizations

```python
# 3D scatter plot
fig = px.scatter_3d(
    Nigerian_suicide_cases,
    x="DIM_AGE",
    y="Suicide_Rate_N",
    z="Gender",  # Use Gender as the third axis (categorical)
    color="Gender",  # Color-code by gender
    title="Suicide Rate by Age and Gender in Nigeria (3D)",
    labels={"Suicide_Rate_N": "Suicide Rate (per 100k)", "Age": "Age (Years)"},
    hover_data={"DIM_AGE": True, "Gender": True, "Suicide_Rate_N": ":.2f"},
    symbol="Gender",  # Optional: Different symbols for genders
    opacity=0.8,
    height=600,
)

fig.update_layout(
    scene=dict(
        xaxis_title="DIM_AGE",
        yaxis_title="Suicide_Rate_N",
        zaxis_title="Gender",
    ),
    margin=dict(l=0, r=0, b=0, t=30),
)

# Show the plot
fig.show()
```

```python
def update_plots(selected_gender):
    filtered_data = data if selected_gender == "All" else Nigerian_suicide_cases[Nigerian_suicide_cases["Gender"] == selected_gender]
    
    # Scatter plot
    scatter = px.scatter(
        filtered_data, 
        x="DIM_AGE", 
        y="Suicide_Rate_N", 
        color="Gender",
        title="Suicide Rate by Age and Gender"
    )
    
    scatter.show()

# Add a gender dropdown
interact(update_plots, selected_gender=widgets.Dropdown(options=["TOTAL", "MALE", "FEMALE"]))
```

```python
# Line plot to show suicide mortality rate for relevant years
```

```python
male = Nigerian_suicide_cases[Nigerian_suicide_cases["Gender"] == "MALE"]
female = Nigerian_suicide_cases[Nigerian_suicide_cases["Gender"] == "FEMALE"]
total = Nigerian_suicide_cases[Nigerian_suicide_cases["Gender"] == "TOTAL"]
```

```python
plt.figure(figsize=(12, 6))
sns.lineplot(
    data=male,
    x="Suicide_Year", y="Suicide_Rate_N",
    marker="o", label="Male", color="#1f77b4"
)
sns.lineplot(
    data=female,
    x="Suicide_Year", y="Suicide_Rate_N",
    marker="o", label="Female", color="#ff7f0e"
)
sns.lineplot(
    data=total,
    x="Suicide_Year", y="Suicide_Rate_N",
    marker="o", label="Total", color="#2ca02c"
)
plt.fill_between(
    male["Suicide_Year"], male["Lower_Rate_NL"], male["Upper_Rate_NU"],
    alpha=0.2, color="#1f77b4"
)
plt.fill_between(
    female["Suicide_Year"], female["Lower_Rate_NL"], female["Upper_Rate_NU"],
    alpha=0.2, color="#ff7f0e"
)
plt.title("Suicide Mortality Rates in Nigeria (2000–2017)")
plt.xlabel("Year")
plt.ylabel("Rate per 100,000")
plt.ylim(0, 12)
plt.grid(alpha=0.3)
plt.legend()
```

```python
# Analysis by Age Group
age_data = Nigerian_suicide_cases[Nigerian_suicide_cases["DIM_AGE"].notna()]

plt.figure(figsize=(12, 6))
sns.boxplot(
    x="DIM_AGE",
    y="Suicide_Rate_N",
    data=age_data,
    palette="Set2"
)
plt.title("Suicide Rates by Age Group")
plt.xlabel("Age Group")
plt.ylabel("Rate per 100,000")
plt.xticks(rotation=45)
plt.show()
```

```python
# Error Bar
Nigerian_suicide_cases = Nigerian_suicide_cases.sort_values('Suicide_Year')
lower_errors = Nigerian_suicide_cases['Suicide_Rate_N'] - Nigerian_suicide_cases['Lower_Rate_NL']
upper_errors = Nigerian_suicide_cases['Upper_Rate_NU'] - Nigerian_suicide_cases['Suicide_Rate_N']
error = [lower_errors, upper_errors]

# Create the error bar plot
plt.figure(figsize=(10, 6))
plt.errorbar(
    Nigerian_suicide_cases['Suicide_Year'],
    Nigerian_suicide_cases['Suicide_Rate_N'],
    yerr=error,
    fmt='o-',  # 'o' for marker, '-' for line
    ecolor='gray',
    capsize=4,
    label='Suicide Rate with Uncertainty'
)

# Labeling
plt.title('Suicide Rate in Nigeria (with Uncertainty)')
plt.xlabel('Suicide Year')
plt.ylabel('Suicide Rate per 100,000')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
